## Problem 6: Factorial (Loop)

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter number: "))
result = factorial(num)
print("Factorial is:", result)
```

## Problem 8: Prime Check

```python
def prime_checker(num):
    if num <= 1:
        print("Not a prime number")
        return

    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            return

    print("It's a prime number")

num = int(input("Enter number: "))
prime_checker(num)
```

## Problem 9: Fibonacci (First N)

```python
# Take input from the user
n = int(input("Enter how many Fibonacci numbers you want: "))

# Initialize the first two Fibonacci numbers
fib_sequence = []

a, b = 0, 1

for i in range(n):
    fib_sequence.append(a)
    a, b = b, a + b  # Update a and b

# Print the Fibonacci sequence
print("The first", n, "Fibonacci numbers are:")
print(fib_sequence)
```

## Problem 15: Manual Sort

```python
# Step 1: Input numbers as a list
lst = list(map(int, input("Enter numbers to sort: ").split()))

# Step 2: Manual Selection Sort
for i in range(len(lst)):
    min_index = i
    for j in range(i+1, len(lst)):
        if lst[j] < lst[min_index]:
            min_index = j
    # Swap the smallest element with current position
    lst[i], lst[min_index] = lst[min_index], lst[i]

# Step 3: Print sorted list space-separated
print(*lst)
```

## Problem 16: Remove Duplicates

```python
lst = list(map(int, input("Enter numbers to sort: ").split()))

unique_lst = []

for item in lst:
    if item not in unique_lst:
        unique_lst.append(item)

print(unique_lst)
```
