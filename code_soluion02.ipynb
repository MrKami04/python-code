## Problem 6: Factorial (Loop)

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter number: "))
result = factorial(num)
print("Factorial is:", result)
```

## Problem 8: Prime Check

```python
def prime_checker(num):
    if num <= 1:
        print("Not a prime number")
        return

    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            return

    print("It's a prime number")

num = int(input("Enter number: "))
prime_checker(num)
```

## Problem 15: Manual Sort

```python
# Step 1: Input numbers as a list
lst = list(map(int, input("Enter numbers to sort: ").split()))

# Step 2: Manual Selection Sort
for i in range(len(lst)):
    min_index = i
    for j in range(i+1, len(lst)):
        if lst[j] < lst[min_index]:
            min_index = j
    # Swap the smallest element with current position
    lst[i], lst[min_index] = lst[min_index], lst[i]

# Step 3: Print sorted list space-separated
print(*lst)
```

## Problem 17: Sum of Digits

```python
num = int(input("Enter a 3-digit number: "))
digits = []

temp = num
while temp > 0:
    digits.append(temp % 10)  # take the last digit
    temp = temp // 10          # remove last digit

digits.reverse()  # optional: put digits in original order

print("Sum of Digits:", sum(digits))

```

## Problem 18: GCD of Two Numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

gcd = 1
small = min(a, b)
for i in range(small, 0, -1):
    if a % i == 0 and b % i == 0:
        gcd = i
        break

print("GCD is:", gcd)
```

```python
# Using Euclidean Algorithm 
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    a, b = b, a % b

print("GCD is:", a)
```

## Problem 19: LCM of Two Numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# Step 1: Find GCD using Euclidean algorithm
x, y = a, b
while y != 0:
    x, y = y, x % y
gcd = x

# Step 2: Find LCM using formula
lcm = (a * b) // gcd

print("LCM is:", lcm)
```

## Problem 25: Primes in Range

```python
L, R = map(int, input("Enter L and R: ").split())

found = False  # Flag to check if any prime exists

for num in range(L, R + 1):
    if num <= 1:
        continue
    # Check if num is prime
    for i in range(2, num):
        if num % i == 0:
            break
    else:  # No break → prime
        print(num, end=" ")
        found = True

if not found:
    print("NA")

        
```

## Problem 26: Armstrong Number

```python
# Step 1: Take input
num = int(input("Enter a number: "))

# Step 2: Store original number for comparison
original_num = num

# Step 3: Find number of digits
n = len(str(num))

# Step 4: Calculate sum of digits raised to the power n
sum_of_powers = 0
while num > 0:
    digit = num % 10           # Get last digit
    sum_of_powers += digit ** n
    num = num // 10            # Remove last digit

# Step 5: Check if sum equals original number
if sum_of_powers == original_num:
    print(original_num, "is an Armstrong number")
else:
    print(original_num, "is not an Armstrong number")
```

## Problem 28: Merge Two Sorted Lists

```python
list1 = list(map(int, input("Enter first sorted list: ").split()))
list2 = list(map(int, input("Enter second sorted list: ").split()))

i = j = 0
merged = []

# Compare elements from both lists
while i < len(list1) and j < len(list2):
    if list1[i] < list2[j]:
        merged.append(list1[i])
        i += 1
    else:
        merged.append(list2[j])
        j += 1

# Add remaining elements (if any)
while i < len(list1):
    merged.append(list1[i])
    i += 1

while j < len(list2):
    merged.append(list2[j])
    j += 1

print(*merged)
```
