```python
num = int(input("Enter rows"))

for i in range(1, num+1):
    for j in range(1,i+1):
        print("*",end = " ")
    
    print()
```

```python
def pyramid(rows):
    for i in range(rows):
        print(' '*(rows-i-1)+'*'*(i+1))
pyramid(5)
```

```python
num = int(input("Enter rows"))
k = 1
for i in range(1, num+1):
    for j in range(1,k+1):
        print("*",end = " ")
    k += 2
    
    print()
```

```python
num = int(input("Enter rows"))

for i in range(0, num):
    for j in range(0,num-i-1):
        print(end=" ")

    for j in range(0,i+1):
        print("*",end = " ")
    
    print()
```

```python
num = int(input("Enter rows"))

for i in range(num,0,-1):
    for j in range(0,num-i):
        print(end=" ")

    for j in range(0,i):
        print("*",end = " ")
    
    print()
```

```python
def pyramid(rows):
    for i in range(rows):
        print(' '*(rows-i-1)+'* '*(i+1))
        
    for j in range(rows-1,0,-1):
        print(' '*(rows-j)+'* '*(j))

    
pyramid(5)
```
