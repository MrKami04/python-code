## Magic / Dunder Methods in Python
#### “Dunder” = Double Under

```python
# Example:
# __init__
# __str__
# __add__
# They are called:
# 👉 Magic Methods
# 👉 Special Methods
# 👉 Dunder Methods
```

### 🟢 1️⃣ __init__ → Constructor

```python
class Person:
    def __init__(self, name):
        self.name = name

p = Person("Kamran")
# Python internally does:
Person.__init__(p, "Kamran")
```

### 🟢 2️⃣ __str__ → User Friendly String

```python
# Controls what prints when you do:
print(object)
```

```python
# Without __str__
class Car:
    pass

c = Car()
print(c)
```

```python
# With __str__
class Car:
    def __str__(self):
        return "This is a Car object"

c = Car()
print(c)
```

### 🟢 3️⃣ __repr__ → Developer Friendly String

```python
# Used for:
object
repr(object)
```

```python
# Goal:
# Give detailed representation for debugging.
# Example:

class Car:
    def __repr__(self):
        return "Car()"
```

### 🟢 4️⃣ __add__ → Operator Overloading

```python
# Controls + operator.
# Example:
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

n1 = Number(10)
n2 = Number(5)

result = n1 + n2
# n1 + n2 actually means:
n1.__add__(n2)   # That is polymorphism with operators.
print(result)
```
