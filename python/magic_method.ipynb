## Magic / Dunder Methods in Python
#### “Dunder” = Double Under

```python
# Example:
# __init__
# __str__
# __add__
# They are called:
# 👉 Magic Methods
# 👉 Special Methods
# 👉 Dunder Methods
```

### 🟢 1️⃣ __init__ → Constructor

```python
class Person:
    def __init__(self, name):
        self.name = name

p = Person("Kamran")
# Python internally does:
Person.__init__(p, "Kamran")
```

### 🟢 2️⃣ __str__ → User Friendly String

```python
# Controls what prints when you do:
print(object)
```

```python
# Without __str__
class Car:
    pass

c = Car()
print(c)
```

```python
# With __str__
class Car:
    def __str__(self):
        return "This is a Car object"

c = Car()
print(c)
```

### 🟢 3️⃣ __repr__ → Developer Friendly String

```python
# Used for:
object
repr(object)
```

```python
# Goal:
# Give detailed representation for debugging.
# Example:

class Car:
    def __repr__(self):
        return "Car()"
```

### 🟢 4️⃣  __add__  → Operator Overloading

```python
# Controls + operator.
# Example:
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

n1 = Number(10)
n2 = Number(5)

result = n1 + n2
# n1 + n2 actually means:
n1.__add__(n2)   # That is polymorphism with operators.
print(result)
```

### 🟢 5️⃣ __len__

```python
class Book:
    def __init__(self, pages):
        self.pages = pages

    def __len__(self):
        return self.pages

b = Book(200)
print(b)
print(len(b))
```

### 🟢 6️⃣ __eq__ → Equality (==)

```python
class Person:
    def __init__(self, name):
        self.name = name

    def __eq__(self, other):
        return self.name == other.name

p1 = Person("Ali")
p2 = Person("Ali")

print(p1 == p2)
```

```python
class BankAccount:

    bank_name = "Python National Bank"   # Class Variable

    def __init__(self, account_number, holder_name, balance=0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.__balance = balance   # Encapsulation (private variable)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Deposit successful.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawal successful.")
        else:
            print("Insufficient balance.")

    def check_balance(self):
        print("Current Balance:", self.__balance)

    def __str__(self):
        return f"Account[{self.account_number}] - {self.holder_name}"
```

```python
acc1 = BankAccount(101, "Kamran", 1000)
acc2 = BankAccount(102, "Ali", 500)

print(acc1)
print(acc2)

acc1.deposit(200)
acc1.withdraw(300)
acc1.check_balance()
```
