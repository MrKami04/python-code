```python
# methods = Actions / Behavior
# Simple:
# 👉 If function is outside class → Normal function
# 👉 If function is inside class → Method
class Student:
    def greet(self):
        print("Hello Student")

s1 = Student()
s1.greet()
```

```python
# Method Using Instance Variable
class Student:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello", self.name)

s1 = Student("Ali")
s2 = Student("Ahmed")

s1.greet()
s2.greet()
```

```python
# Method With Parameters
class Calculator:
    def add(self, a, b):
        print(a + b)

c1 = Calculator()
c1.add(5, 3)
```

### Class Methods vs Static Methods


```python
## Class Method
### ✅ What is a Class Method? (use cls instaed of self)
#### A method that works with:
#### Class variables
#### Class itself
```

```python
class Student:
    school = "ABC School"   # Class variable

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_school(cls, new_name):
        cls.school = new_name

s1 = Student("Ali")
s2 = Student("Ahmed")

print(s1.school)
Student.change_school("XYZ School")
print(s2.school)
```

```python
## Static Method
###✅ What is a Static Method?
#### A method that:
#### Does NOT use self
#### Does NOT use cls
#### Is just inside class for logical grouping
```

```python
### Static method:
### Does not access instance variables
### Does not access class variables
### Works like normal function
### Just placed inside class for organization
```

```python
class MathUtils:

    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.add(5, 3))
```

```python
class Car:
    wheels = 4
    print("Old wheels is :",wheels)
    def __init__(self,brand):
        self.brand = brand
    def show_brand(self):
        print("Car brand is",self.brand)
    @classmethod
    def change_wheels(cls,new_wheel):
        cls.new_wheel = new_wheel
        print("New wheels is :",new_wheel)

    @staticmethod
    def is_fast(speed):
        return speed > 100
        
c1 = Car("Toyota")
c1.show_brand()
c1.change_wheels(5)
c1.is_fast(110)
```
