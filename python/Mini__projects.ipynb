```python
class Car:

    def __init__(self, color, speed):
        self.color = color
        self.speed = speed

    def start(self):
        print(f"{self.color} car is starting at {self.speed} km/h.")

    def stop(self):
        print(f"{self.color} car has stopped.")

BMW = Car("black", 120)
Toyota = Car("white", 140)
Honda = Car("blue", 150)

print(BMW.color, BMW.speed)
print(Toyota.color, Toyota.speed)
print(Honda.color, Honda.speed)
print("---------------")
BMW.start()
Honda.stop()
```
