### Class Methods vs Static Methods


```python
## Class Method
### ✅ What is a Class Method? (use cls instaed of self)
#### A method that works with:
#### Class variables
#### Class itself
```

```python
class Student:
    school = "ABC School"   # Class variable

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_school(cls, new_name):
        cls.school = new_name

s1 = Student("Ali")
s2 = Student("Ahmed")

print(s1.school)
Student.change_school("XYZ School")
print(s2.school)
```

```python
## Static Method
###✅ What is a Static Method?
#### A method that:
#### Does NOT use self
#### Does NOT use cls
#### Is just inside class for logical grouping
```

```python
### Static method:
### Does not access instance variables
### Does not access class variables
### Works like normal function
### Just placed inside class for organization
```

```python
class MathUtils:

    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.add(5, 3))
```

## 🧠 What is Encapsulation?
### Encapsulation means:
#### 👉 Wrapping data (variables) and methods together
#### 👉 Restricting direct access to data
#### Simple words: Protecting object data from direct access.

```python
# Problem Without Encapsulation
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

account = BankAccount(1000)
account.balance = -5000   # Wrong!
print(account.balance)
```

### Data Hiding
#### Python gives special way to hide data.
#### __variable : Double underscore before variable name.

### Real World Meaning
#### Encapsulation =
##### ✔ Data protection
##### ✔ Control access
##### ✔ Prevent wrong changes
##### ✔ Secure object state

```python
# Example With Encapsulation
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient Balance")

    def get_balance(self):
        print(id(self.__balance))
        return self.__balance
        

account = BankAccount(1000)
account.deposit(500)
account.withdraw(3000)

print(account.get_balance())
```

```python
## 🧠 What is Abstraction?
### Abstraction means:
#### 👉 Hiding implementation details
#### 👉 Showing only necessary features
#### Simple words: Show what is important, hide how it works.
```

```python
#### 🔵 In Python — How Do We Do Abstraction?
#### Two ways:
#### 1️⃣ Using private variables (basic abstraction)
#### 2️⃣ Using Abstract Classes (advanced abstraction)
```

```python
# Example 1 — Basic Abstraction
class Car:
    def start(self):
        print("Car started")

    def drive(self):
        print("Car is moving")

c1 = Car()
c1.start()
c1.drive()
```

```python
# Abstract Class (Professional Level)
# Python provides module:
# from abc import ABC, abstractmethod
```

```python
# Example 2 — Real Abstraction
from abc import ABC, abstractmethod

class Animal(ABC):   # Abstract class

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

d = Dog()
d.sound()
```

```python
# Why We Use Abstract Class?
# To:
# ✔ Force subclasses to implement certain methods
# ✔ Create common structure
# ✔ Hide base implementation
```

# Mini project

```python
# from abc import ABC, abstractmethod
import math

# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


# Rectangle Class
class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


# Circle Class
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2


# Testing
rect = Rectangle(10, 5)
circle = Circle(7)

print("Rectangle Area:", rect.area())
print("Circle Area:", circle.area())
```

## 🧠 What is Inheritance?
#### Inheritance means:
##### 👉 One class can use properties and methods of another class.
##### 👉 Child class gets features of Parent class.
##### Simple:Reuse code instead of writing again.

### Example 1 — Simple Inheritance

```python
class Animal:
    def eat(self):
        print("Animal is eating")

class Dog(Animal):
    def bark(self):
        print("Dog is barking")

d = Dog()
d.eat()
d.bark()
```

### Example 2 — With Constructor
##### Now important part.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def show(self):
        print("Name:", self.name)

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)   # Call parent constructor
        self.grade = grade

    def display(self):
        print("Grade:", self.grade)

s1 = Student("Ali", "A")
s1.show()
s1.display()
```

```python
### Important: super()
# super().__init__(name) #This calls parent constructor.
# Without it:Parent variables will not initialize.
```

## 🎯 Types of Inheritance in Python

### 1️⃣ Single Inheritance
### 2️⃣ Multiple Inheritance
### 3️⃣ Multilevel Inheritance
### 4️⃣ Hierarchical Inheritance
### 5️⃣ Hybrid Inheritance

### 1️⃣ Single Inheritance
##### ✅ Definition: One child class inherits from one parent class.

```python
class Animal:
    def eat(self):
        print("Animal is eating")

class Dog(Animal):
    def bark(self):
        print("Dog is barking")

d = Dog()
d.eat()
d.bark()
```

### 2️⃣ Multiple Inheritance
##### ✅ Definition: One child class inherits from more than one parent class.

```python
class Father:
    def skills(self):
        print("Gardening")

class Mother:
    def talent(self):
        print("Cooking")

class Child(Father, Mother):
    def hobby(self):
        print("Playing")

c = Child()
c.skills()
c.talent()
c.hobby()
```

### 3️⃣ Multilevel Inheritance
##### ✅ Definition: Inheritance in multiple levels (grandparent → parent → child).

```python
class Grandfather:
    def property(self):
        print("Land")

class Father(Grandfather):
    def house(self):
        print("House")

class Son(Father):
    def car(self):
        print("Car")

s = Son()
s.property()
s.house()
s.car()
```

### 4️⃣ Hierarchical Inheritance
##### ✅ Definition: Multiple child classes inherit from one parent class.

```python
class Vehicle:
    def start(self):
        print("Vehicle started")

class Car(Vehicle):
    def drive(self):
        print("Car driving")

class Bike(Vehicle):
    def ride(self):
        print("Bike riding")

c = Car()
b = Bike()

c.start()
b.start()
```

### 5️⃣ Hybrid Inheritance
##### ✅ Definition: Combination of multiple types of inheritance.
#### Example: Multiple + Multilevel mixed together.

```python
class A:
    def method_A(self):
        print("Class A")

class B(A):
    def method_B(self):
        print("Class B")

class C(A):
    def method_C(self):
        print("Class C")

class D(B, C):
    def method_D(self):
        print("Class D")

obj = D()

obj.method_A()
obj.method_B()
obj.method_C()
obj.method_D()
```

### 🧠 What is Method Overriding?
#### Method Overriding means:
##### 👉 Child class changes the method of parent class
##### 👉 Same method name
##### 👉 Different implementation
### Simple: Child replaces parent behavior.

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")
    # pass

d = Dog()
d.sound()
```

### Python checks:
##### 1️⃣ Is sound() in Dog? → YES ✅
##### 2️⃣ So use Dog’s version
##### 3️⃣ Ignore Animal’s version

### Very Important Rule
#### For overriding:
##### Method name must be same 
##### Parameters must be same
##### Child class must inherit from parent

### 🔵 Using super() in Overriding
#### Sometimes we want:
##### ✔ Use parent method
##### ✔ AND add extra behavior
#### Example:

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Dog barks")

d = Dog()
d.sound()
```

#### 📌 In One Line
##### Method Overriding = Child replaces parent method.

## MRO (Method Resolution Order)
### 🧠 What is MRO?
#### MRO tells Python:
##### 👉 When multiple parent classes exist,
##### 👉 Which class method should be executed first?
#### Simple: MRO decides the order in which Python searches for methods.

### 🔵 Example 1 — Simple Multiple Inheritance

```python
class A:
    def show(self):
        print("Class A")

class B:
    def show(self):
        print("Class B")

class C(A, B):
    pass

obj = C()
obj.show()
```

###  How To Check MRO?

```python
print(C.__mro__)
print(C.mro())
```

```python
# Python uses:👉 C3 Linearization Algorithm
```

### 🔵 Example 2 — Multilevel + Multiple (Complex Case)

```python
class A:
    def show(self):
        print("A")

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

obj = D()
obj.show()
```

```python
# Python MRO becomes: D → B → C → A → object
```

### 🔥 Example With Overriding

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass

obj = D()
obj.show()
# D → B → C → A → object
```

```python
# 🧠 Golden Rule To Remember
#  Python checks:
# 1️⃣ Current class
# 2️⃣ Parent classes (left to right)
# 3️⃣ Moves upward
# 4️⃣ Stops at first match
```

### What is Polymorphism?
##### 👉 One name
##### 👉 Many forms
#### Simple definition:
##### Same method name
##### Different behavior

### 🔵 Types of Polymorphism in Python
#### There are mainly:
##### 1️⃣ Method Overriding (Runtime Polymorphism)
##### 2️⃣ Operator Overloading
##### 3️⃣ Duck Typing

```python
# 1️⃣ Method Overriding (Runtime Polymorphism)
# We already studied this.

class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

animals = [Dog(), Cat()]

for a in animals:
    a.sound()
```

```python
# 2️⃣ Operator Overloading
# Same operator works differently.
# Example:

print(5 + 3)        # Addition
print("Hi " + "Kamran")  # String concatenation
```

```python
# 3️⃣ Duck Typing (Very Pythonic 🐍)
# Rule:
# 👉 “If it walks like a duck and talks like a duck, it’s a duck.”
# Means:
# Object type doesn’t matter
# Method existence matters
# Example:

class Dog:
    def sound(self):
        print("Dog barks")

class Cat:
    def sound(self):
        print("Cat meows")

def make_sound(animal):
    animal.sound()

make_sound(Dog())
make_sound(Cat())
```
