## 10. Take a character and check whether it’s uppercase, lowercase, a digit, or a special character.

```python
# Take input from the user
char = input("Enter a single character: ")

# Ensure only one character is entered
if len(char) != 1:
    print("Please enter exactly one character.")
else:
    if char.isupper():
        print("The character is uppercase.")
    elif char.islower():
        print("The character is lowercase.")
    elif char.isdigit():
        print("The character is a digit.")
    else:
        print("The character is a special character.")
```

## 4. Check if one of two given numbers is a multiple of the other. 

```python
num1, num2 = map(int, input("Ener numbers:").split( ))
# note: enter two numbers
numbers = [num1, num2]
max_num = max(numbers)
min_num = min(numbers)

if max_num % min_num == 0:
    print("Yes number is a multiple of other.")
else:
    print("No, number is not a multiple of other.")
```

## 5. Take the hour of the day (0–23) and print “Good Morning”, “Good Afternoon”, “Good Evening”, or “Good Night”. 

```python
# Input hour from the user
hour = int(input("Enter the hour (0-23): "))

# Check the time and print the greeting
if 5 <= hour < 12:
    print("Good Morning")
elif 12 <= hour < 17:
    print("Good Afternoon")
elif 17 <= hour < 21:
    print("Good Evening")
elif (21 <= hour <= 23) or (0 <= hour < 5):
    print("Good Night")
else:
    print("Invalid hour! Please enter 0-23.")
```

## 8. Take an alphabet character and check if it lies between ‘a’ and ‘m’ or ‘n’ and ‘z’. 

```python
#  Characters in Python have ASCII/Unicode values
ch = input("Enter an alphabet character: ").lower()

if 'a' <= ch <= 'm':
    print("The character lies between 'a' and 'm'")
elif 'n' <= ch <= 'z':
    print("The character lies between 'n' and 'z'")
else:
    print("Invalid input (not an alphabet)")
```

## 9. Take a day number (1–7) and print the corresponding day name. 

```python
day = int(input("Enter day number: "))

match day :
    case 1:
        print("Sunday")
    case 2:
        print("Monday")
    case 3:
        print("Tuesday")
    case 4:
        print("Wednesday")
    case 5:
        print("Thursday")
    case 6:
        print("Friday")
    case 7:
        print("Saturday")
    case _:
        print("invalid")
```

## 10. Take a month number (1–12) and print the number of days in that month (ignore leap years).
