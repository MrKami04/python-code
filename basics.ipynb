```python
age = 16

if age >= 18:
    print("You are eligible for vote")
else:
    print("You cannot eligible for vote.")
```

### 03 if-elif-else condition:

```python
marks = 75

if marks >= 90:
    print("Your grade is A")
elif marks >= 70:
    print("Your grade is B")
elif marks >= 50:
    print("Your grade is C")
else:
    print("You are fail")
```

### 4 Nested if statement: if statement inside of the if statement.

```python
age = 18
citizen = True

if age >= 18:
    if citizen:
        print("You are eligible to vote")
    else:
        print("You must be citizen")
else:
    print("You are underage.")
```

# Real-Life Example (ATM System)

```python
pin = 1234
entered_pin = int(input("Enter pin :"))

if entered_pin == pin:
    amount = int(input("Enter your amount :"))
    if amount <= 5000:
        entered_pin = int(input("Enter pin :"))
        if entered_pin == pin:
            print("Please! collect your amount",amount)
        else:
            print("Invalid pin!")
    else:
        print("Enter amount below 5000.")
else:
    print("Invalid pin!")
    
```

### 01 while loop:

```python
i = 0
while i <= 5:
    print(i)
    i += 1
    
```

### 02 break statement

```python
week_days = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]

i = 0
while i < len(week_days):
    if week_days[i] == "Friday":
        break
    print(week_days[i])
    i += 1
```

### 01 Function: A function is a block of code that performs a specific task and can be reused.

### 02 Doc String (Documentation String)
#### A docstring explains what a function does.

```python
def add():
    """This function adds two numbers"""
    print(10 + 20)

print(add.__doc__)
```

### 03 Function Input Arguments
#### Arguments allow passing data to a function.

```python
def add(a, b):
    print(a + b)

add(5, 3)
```

### 04 Order of Input Arguments
#### Arguments are passed in the same order as parameters.

```python
def show(a, b):
    print(a, b)

show(10, 20)   # Correct order
show(20, 10)   # Different output
```

### 05 Return Statement
#### Returns a value from a function.

```python
def square(num):
    return num * num

result = square(4)
print(result)
```

### 07 *args in Function
#### Accepts multiple positional arguments.

```python
def total(*args):
    sum = 0
    for i in args:
        sum += i
    return sum

print(total(10, 20))
print(total(1, 2, 3, 4, 5))
```

### 08 **kwargs in Function
#### Accepts multiple keyword arguments (key=value).

```python
def student(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

student(name="Ali", age=20, course="Python")
```

### 09 *args and **kwargs Together


```python
def display(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

display(1, 2, 3, name="Sara", city="Lahore")
```

### 10 Lambda Function

```python
add = lambda a, b: a + b
print(add(3, 4))
```
